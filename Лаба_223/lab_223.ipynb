## 1) Синий свет

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
%matplotlib inline


y1 = [21.5, 29.5, 35, 39.5, 43.5, 47.2, 50.7, 54, 56.5, 60]
y2 = [26, 33, 37, 42, 45.8, 49, 52, 55, 58, 60.9]

y1_2 = [num ** 2 for num in y1]
y2_2 = [num ** 2 for num in y2]

x_blue = np.linspace(0, 10, len(y1))

fig, axs = plt.subplots(1, 2, figsize=(14, 6))
ax1, ax2 = axs

ax1.plot(x_blue, y1_2, marker = 'o', linestyle = 'none')
ax1.set_title('Синий (светлые кольца)')
ax1.set_xlabel('номер кольца')
ax1.set_ylabel('мм * 10^(-4)')
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.minorticks_on()
ax1.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

x_line1 = np.linspace(0, 10, 10)  
y_line1 = 0.0302 * x_line1 * 10000 + 515      
ax1.plot(x_line1, y_line1, 'r-')

ax2.plot(x_blue, y2_2, marker = 'o', linestyle = 'none')
ax2.set_title('Синий (темные кольца)')
ax2.set_xlabel('номер кольца')
ax2.set_ylabel('мм * 10^(-4)')
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.minorticks_on()
ax2.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

x_line2 = np.linspace(0, 10, 10)  
y_line2 = 0.0303 * x_line2 * 10000 + 700
ax2.plot(x_line2, y_line2, 'r-')

plt.show()
```

## 2) Зеленый свет

```python
y3 = [20, 28, 34.4, 39.5, 44.5, 48.8, 52.4, 55.7, 58.6, 62.2, 65.2, 68.2, 70.8, 73.5, 75.9, 78.2, 80.8, 83.2, 85.2, 87.7]
y4 = [25.5, 32, 37, 42, 46.5, 50.5, 53.2, 57.8, 60.8, 63.8, 66.9, 69.3, 72, 74.9, 77, 79.5, 82, 84.5, 86.5, 88.8]

y3_2 = [num ** 2 for num in y3]
y4_2 = [num ** 2 for num in y4]

x_green = np.linspace(0, 20, len(y4))

fig, axs = plt.subplots(1, 2, figsize = (14, 6))
ax1, ax2 = axs

ax1.plot(x_green, y3_2, marker = 'o', color = 'green', linestyle = 'none')
ax1.set_title('Зеленый (светлые кольца)')
ax1.set_xlabel('номер кольца')
ax1.set_ylabel('мм * 10^(-4)')
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.minorticks_on()
ax1.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

x_line3 = np.linspace(0, 20, 20)  
y_line3 = 0.0368 * x_line3 * 10000 + 398 
ax1.plot(x_line3, y_line3, 'r-')

ax2.plot(x_green, y4_2, marker = 'o', color = 'green',linestyle = 'none')
ax2.set_title('Зеленый (темные кольца)')
ax2.set_xlabel('номер кольца')
ax2.set_ylabel('мм * 10^(-4)')
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.minorticks_on()
ax2.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

x_line4 = np.linspace(0, 20, 20)  
y_line4 = 0.03625 * x_line4 * 10000 + 603
ax2.plot(x_line4, y_line4, 'r-')

plt.show()
```
